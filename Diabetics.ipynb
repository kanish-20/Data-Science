```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

```

```python
df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
X=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
Y=df['outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
sns.scatterplot(x='Insulin', y='outcome', data=df)
plt.title('Diabetes')
plt.xlabel('Insulin')
plt.ylabel('outcome')
plt.show()


print("Correlation:", df['Insulin'].corr(df['outcome']))
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
Diabetes=model.predict(X_test)
```

```python
print("MSE:",mean_squared_error(Y_test,Diabetes))
print("r2:",r2_score(Y_test,Diabetes))
```

```python
print("Diabetes:",Diabetes)
```
